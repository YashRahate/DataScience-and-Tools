```python
# Create dummy variables for categorical columns
df_dummies = pd.get_dummies(df, columns=['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years'], drop_first=True)

# Check the resulting dataframe with dummy variables
df_dummies.head()
```

 Product_Category might be unordered so coverting into dummy variable.(Whereas occupation is ordered)

```python
# Create a list of all possible product categories (from 0 to 20)
categories = range(21)

# Loop through each category and create a new column for each category (Product_Category_x)
for category in categories:
    df_dummies[f'Product_Category_{category}'] = ((df_dummies['Product_Category_1'] == category) |
                                                  (df_dummies['Product_Category_2'] == category) |
                                                  (df_dummies['Product_Category_3'] == category)).astype(int)

# Now, df_dummies will have both the dummy variables for categorical columns and the binary product category columns
```

# 7. Standardization and Normalization of columns

```python
df_dummies.describe()
```

**Columns to Standardize**

Purchase: This is a continuous variable with a large range (min: 12, max: 23961), so it should be standardized to have a mean of 0 and a standard deviation of 1.

**Columns to Normalize**
Categorical Columns (like Occupation, gender, Marital_Status,Stay_In_Current_City_Years_* , Product_Category_* columns): These are already in binary form (0 or 1) after creating the dummy variables, so they do not need normalization.

```python
from sklearn.preprocessing import StandardScaler

# Standardize 'Purchase'
scaler = StandardScaler()
df_dummies['Purchase'] = scaler.fit_transform(df_dummies[['Purchase']])
df_dummies.head()
```
