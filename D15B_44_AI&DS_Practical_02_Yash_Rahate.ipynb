```python
import pandas as pd

# Assuming df_dummies already exists
# Step 1: Define the range of categories
categories = range(21)  # Categories from 0 to 20

# Step 2: Initialize an empty DataFrame with required columns
df_dummies1 = pd.DataFrame(0, index=df_dummies.index, columns=[f'category_{i}' for i in categories])

# Step 3: Iterate over Product_Category columns and update df_dummies1
for col in ['Product_Category_1', 'Product_Category_2', 'Product_Category_3']:
    for category in categories:
        df_dummies1[f'category_{category}'] |= (df_dummies[col] == category).astype(int)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("Data before removing outliers (df_2):", df_2.shape)

# Visualize outliers with a box plot for df_2
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_2['Purchase'])
plt.title('Box Plot of Purchase Amount (df_2)')
plt.show()

# Calculate IQR for df_2
Q1 = df_2['Purchase'].quantile(0.25)
Q3 = df_2['Purchase'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds for df_2
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Filter out outliers from df_2
df_2_no_outliers = df_2[(df_2['Purchase'] >= lower_bound) & (df_2['Purchase'] <= upper_bound)]


```

```python
print("Data after removing outliers (df_2_no_outliers):", df_2_no_outliers.shape)

# Visualize the updated box plot for df_2_no_outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_2_no_outliers['Purchase'])
plt.title('Box Plot of Purchase Amount After Outlier Removal (df_2)')
plt.show()
```
