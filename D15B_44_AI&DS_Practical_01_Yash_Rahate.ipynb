# 5. Take care of missing data.

Lets assume 0 to be representing 'no-category'.Hence, fill missing values in Product_Category_2 and Product_Category_3 with 0 .

```python
# Fill missing values in Product_Category_2 and Product_Category_3 with 0
df['Product_Category_2'] = df['Product_Category_2'].fillna(0)
df['Product_Category_3'] = df['Product_Category_3'].fillna(0)

# Convert Product_Category_2 and Product_Category_3 to integer type
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)

# Verify the changes
df[['Product_Category_2', 'Product_Category_3']].head()
```

```python
# Create dummy variables for categorical columns
df_dummies = pd.get_dummies(df, columns=['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years'], drop_first=True)

# Convert only boolean columns to integers
for col in df_dummies.select_dtypes(include=['bool']).columns:
    df_dummies[col] = df_dummies[col].astype(int)

# Check the updated dataframe
df_dummies.head()
```

 Product_Category might be unordered so coverting into dummy variable.(Whereas occupation is ordered)

```python
df_dummies.info()
```

```python
import pandas as pd

# Assuming df_dummies already exists
# Step 1: Define the range of categories
categories = range(21)  # Categories from 0 to 20

# Step 2: Initialize an empty DataFrame with required columns
df_dummies1 = pd.DataFrame(0, index=df_dummies.index, columns=[f'category_{i}' for i in categories])

# Step 3: Iterate over Product_Category columns and update df_dummies1
for col in ['Product_Category_1', 'Product_Category_2', 'Product_Category_3']:
    for category in categories:
        df_dummies1[f'category_{category}'] |= (df_dummies[col] == category).astype(int)
```
